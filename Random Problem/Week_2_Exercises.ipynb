### 4.10: (GUESS THE NUMBER) Write a script that plays “guess the number.” Choose the number to be guessed by selecting a random integer in the range 1 to 1000. Do not reveal this number to the user. Display the prompt "Guess my number between 1 and 1000 with the fewest guesses:". The player inputs a first guess. If the guess is incorrect, display "Too high. Try again." or "Too low. Try again." as appropriate to help the player “zero in” on the correct answer, then prompt the user for the next guess. When the user enters the correct answer, display "Congratulations. You guessed the number!", and allow the user to choose whether to play again.

```python
import random

def guess_the_number():
    # Generate a random number between 1 and 1000
    number_to_guess = random.randint(1, 1000)

    print("Guess my number between 1 and 1000 with the fewest guesses:")

    # Initialize the number of attempts
    attempts = 0

    while True:
        # Increment the attempts count
        attempts += 1

        # Get user input and convert it to an integer
        try:
            guess = int(input("Enter your guess: "))
        except ValueError:
            print("Please enter a valid integer.")
            continue

        # Check if the guess is too high, too low, or correct
        if guess > number_to_guess:
            print("Too high. Try again.")
        elif guess < number_to_guess:
            print("Too low. Try again.")
        else:
            print(f"Congratulations! You guessed the number in {attempts} attempts!")
            break

def play_game():
    while True:
        # Call the guess_the_number function to start the game
        guess_the_number()

        # Ask if the user wants to play again
        play_again = input("Do you want to play again? (yes/no): ").lower()

        if play_again != 'yes':
            print("Thank you for playing! Goodbye.")
            break

# Start the game
play_game()
```

### Display the elements in tabular format. Use the column indices as headings across the top, and the row indices to the left of each row.

```python
# Initialize a 2-by-3 list with zeros
matrix = [[0 for _ in range(3)] for _ in range(2)]

# Set each element's value to the order in which it was processed
count = 1
for i in range(2):
    for j in range(3):
        matrix[i][j] = count
        count += 1

# Print the matrix in tabular format
print("    0  1  2")  # Column headings
for i in range(2):
    # Print row index and row elements
    print(f"{i}   {' '.join(map(str, matrix[i]))}")
```

### 6.5: (COUNTING DUPLICATE WORDS) Write a script that uses a dictionary to determine and print the number of duplicate words in a sentence. Treat uppercase and lowercase letters the same and assume there is no punctuation in the sentence. Use the techniques you learned in Section 6.2.7. Words with counts larger than 1 have duplicates.

```python
# Function to count duplicate words
def count_duplicates(sentence):
    # Convert the sentence to lowercase and split it into words
    words = sentence.lower().split()

    # Create a dictionary to count the occurrences of each word
    word_count = {}

    # Count the occurrences of each word
    for word in words:
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1

    # Find and print the duplicate words
    duplicates = {word: count for word, count in word_count.items() if count > 1}

    if duplicates:
        print("Duplicate words and their counts:")
        for word, count in duplicates.items():
            print(f"'{word}': {count} times")
    else:
        print("No duplicate words found.")

# Example usage
sentence = input("Enter a sentence: ")
count_duplicates(sentence)
```

### 6.7: (CHARACTER COUNTS) Recall that strings are sequences of characters. Use techniques similar to Fig.6.2  to write a script that inputs a sentence from the user, then uses a dictionary to summarize the number of occurrences of each letter. Ignore case, ignore blanks and assume the user does not enter any punctuation. Display a two-column table of the letters and their counts with the letters in sorted order. Challenge: Use a set operation to determine which letters of the alphabet were not in the original string.

```python
import string

def character_counts(sentence):
    # Initialize an empty dictionary for counting letter occurrences
    letter_counts = {}

    # Convert sentence to lowercase and iterate over each character
    for char in sentence.lower():
        if char.isalpha():  # Only count letters, ignore spaces and non-alphabetic characters
            letter_counts[char] = letter_counts.get(char, 0) + 1

    # Display the counts in a sorted order by letter
    print(f"{'Letter':<6}{'Count'}")
    print("-" * 12)
    for letter in sorted(letter_counts):
        print(f"{letter:<6}{letter_counts[letter]}")

    # Determine which letters of the alphabet were not in the sentence
    all_letters = set(string.ascii_lowercase)  # Set of all lowercase letters
    sentence_letters = set(letter_counts.keys())  # Set of letters found in the sentence
    missing_letters = all_letters - sentence_letters  # Set difference to find missing letters

    # Display missing letters
    print("\nMissing letters from the alphabet:")
    print(' '.join(sorted(missing_letters)))

# Input sentence from the user
sentence = input("Enter a sentence: ")
character_counts(sentence)
```

### 6.8: (CHALLENGE: WRITING THE WORD EQUIVALENT OF A CHECK AMOUNT) In check-writing systems, it’s crucial to prevent alteration of check amounts. One common security method requires that the amount be written in numbers and spelled out in words as well. Even if someone can alter the numerical amount of the check, it’s tough to change the amount in words. Create a dictionary that maps numbers to their corresponding word equivalents. Write a script that inputs a numeric check amount that’s less than 1000 and uses the dictionary to write the word equivalent of the amount. For example, the amount 112.43 should be written as:

image.png

```python
# Dictionary mapping numbers to words
number_to_word = {
    0: "ZERO", 1: "ONE", 2: "TWO", 3: "THREE", 4: "FOUR", 5: "FIVE", 6: "SIX", 7: "SEVEN", 8: "EIGHT", 9: "NINE",
    10: "TEN", 11: "ELEVEN", 12: "TWELVE", 13: "THIRTEEN", 14: "FOURTEEN", 15: "FIFTEEN", 16: "SIXTEEN", 17: "SEVENTEEN",
    18: "EIGHTEEN", 19: "NINETEEN", 20: "TWENTY", 30: "THIRTY", 40: "FORTY", 50: "FIFTY", 60: "SIXTY", 70: "SEVENTY",
    80: "EIGHTY", 90: "NINETY"
}

# Helper function to convert numbers less than 1000 into words
def convert_number_to_words(n):
    if n == 0:
        return "ZERO"

    if n < 20:
        return number_to_word[n]

    if n < 100:
        tens, below_ten = divmod(n, 10)
        return number_to_word[tens * 10] + (f" {number_to_word[below_ten]}" if below_ten else "")

    if n < 1000:
        hundreds, remainder = divmod(n, 100)
        if remainder == 0:
            return f"{number_to_word[hundreds]} HUNDRED"
        else:
            return f"{number_to_word[hundreds]} HUNDRED {convert_number_to_words(remainder)}"

# Main function to process check amount
def check_amount_in_words(amount):
    # Split the amount into dollars and cents
    dollars, cents = map(int, f"{amount:.2f}".split('.'))

    # Convert dollars to words
    if dollars > 0:
        dollars_in_words = convert_number_to_words(dollars)
    else:
        dollars_in_words = "ZERO"

    # Format the cents
    cents_in_words = f"{cents}/100"

    # Return the final check amount in words
    return f"{dollars_in_words} AND {cents_in_words}"

# Test the function with input from the user
check_amount = float(input("Enter a check amount less than 1000: "))
print(check_amount_in_words(check_amount))
```

```python
# Initial sets
set1 = {'red', 'green', 'blue'}
set2 = {'cyan', 'green', 'blue', 'magenta', 'red'}

# (a) Union of the two sets (all unique elements from both sets)
union_result = set1.union(set2)
print("Union:", union_result)

# (b) Intersection of the two sets (common elements between both sets)
intersection_result = set1.intersection(set2)
print("Intersection:", intersection_result)

# (c) Difference of set1 from set2 (elements in set1 but not in set2)
difference_result = set1.difference(set2)
print("Difference (set1 - set2):", difference_result)

# (d) Difference of set2 from set1 (elements in set2 but not in set1)
difference_result2 = set2.difference(set1)
print("Difference (set2 - set1):", difference_result2)

# (e) Symmetric difference (elements in either set1 or set2, but not both)
symmetric_difference_result = set1.symmetric_difference(set2)
print("Symmetric Difference:", symmetric_difference_result)

# (f) Check if set1 is a subset of set2
is_subset = set1.issubset(set2)
print("Is set1 a subset of set2?:", is_subset)

# (g) Check if set2 is a superset of set1
is_superset = set2.issuperset(set1)
print("Is set2 a superset of set1?:", is_superset)

# (h) Check if set1 and set2 are disjoint (no common elements)
are_disjoint = set1.isdisjoint(set2)
print("Are set1 and set2 disjoint?:", are_disjoint)
```
